Realizar un escaneo a un IP de su red y a un IP externo (red pública) usando

entre estos dos scanning? Mostrar los resultados y su diferencia.

```python
import subprocess
import os
from datetime import datetime

# Definir las variables para el comando nmap
ip_address = "192.168.0.1/24" 
scan_types = ["-sn"]

# Crear la carpeta para almacenar los resultados si no existe
folder_name = "nmap_results"
os.makedirs(folder_name, exist_ok=True)

# Función para formatear el nombre del archivo basado en el tipo de escaneo y la fecha actual
def get_output_filename(scan_type):
    date_str = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
    return os.path.join(folder_name, f"nmap_output_{scan_type.strip('-')}_{date_str}.xml")

# Ejecutar los escaneos y guardar cada resultado en un archivo separado
for scan_type in scan_types:
    output_file = get_output_filename(scan_type)
    output_format = f" -oX {output_file}"
    
    
    # Formatear el comando usando f-strings
    command = f"nmap {scan_type} {ip_address}{output_format}"

    # Ejecutar el comando nmap
    result = subprocess.run(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

    # Imprimir la salida y los errores
    print("Salida:", result.stdout)
    if result.stderr:
        print("Error:", result.stderr)
```

### Análisis de los Resultados de los Escaneos y Sus Implicaciones en Seguridad de Sistemas:

1. **Escaneo de Red Local (192.168.0.x)**
   - **Detección de Pocos Dispositivos**: Esto podría sugerir una red bien administrada y segura con un control estricto sobre los dispositivos conectados. Sin embargo, también se debe considerar la posibilidad de que dispositivos más sofisticados puedan estar utilizando técnicas para evadir la detección.
   - **Presencia de Dispositivos con Fabricante Identificado y Otros Desconocidos**: Esto podría indicar una mezcla de hardware de confianza y otros potencialmente riesgosos o desconocidos que podrían necesitar investigaciones adicionales o medidas de seguridad específicas.

2. **Escaneo de Red Externa (200.14.249.x)**
   - **Alto Número de Hosts Activos**: La alta cantidad de dispositivos activos en una red externa puede ser un indicativo de una mayor superficie de ataque. Cada host activo podría potencialmente ser un vector para ataques si no está adecuadamente protegido.
   - **Consistencia en las Respuestas de Latencia**: La uniformidad en la latencia entre los hosts podría sugerir una red bien configurada, pero también subraya la necesidad de protección uniforme a través de todos los dispositivos para evitar puntos débiles.

3. **Escaneo a Dominio Específico (informatica.uv.cl)**
   - **Baja Latencia y Respuesta Única**: El hecho de que un dominio específico tenga baja latencia y solo una dirección IP responda puede indicar una configuración eficiente y posiblemente una menor complejidad de red, lo cual podría simplificar la gestión de seguridad. Sin embargo, esto también puede implicar un punto único de fallo, donde un ataque exitoso podría comprometer rápidamente toda la red asociada a ese dominio.

### Conclusiones e Inferencias para la Ciberseguridad

- **Gestión de Riesgos y Seguridad de Red**:
  - Las redes con menor cantidad de dispositivos y baja latencia, como en la red local, pueden ser más fáciles de monitorear y proteger, pero requieren vigilancia constante para detectar cualquier intento de evasión.
  - En contraste, redes con muchos dispositivos y actividad, como la red externa escaneada, necesitan estrategias de segmentación robustas, políticas de seguridad estrictas, y herramientas avanzadas para gestionar la seguridad a gran escala.

- **Planificación de Seguridad Basada en Evidencia**:

Por ejemplo, los dispositivos con fabricantes desconocidos o que no se pueden identificar necesitan más atención y posiblemente políticas más restrictivas.

- **Estrategias Proactivas de Ciberseguridad**:
  - La implementación de medidas proactivas, como la monitorización de la red en tiempo real y las evaluaciones regulares de seguridad, son esenciales para mantener la integridad de tanto redes internas como externas.
