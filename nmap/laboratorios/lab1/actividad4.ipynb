Repita el escaneo de la pregunta 3 pero desde la red pública, ¿que diferencia 
aprecia?, ¿que conclusiones en materia de seguridad puede entregar?

```python

import subprocess
import os
from datetime import datetime

# Definir las variables para el comando nmap
scan_types = ["-sS"]  # Realiza un escaneo SYN Stealth
scan_service = "-sV"  # -sV para detectar los servicios
timing_template = "-T4"  # Establece la velocidad del escaneo a "agresiva"
advanced_options = "-A"  # Habilita la detección de OS, versión de servicio y script
verbosity_level = "-v"  # Incrementa la verbosidad del output
target = "informatica.uv.cl"  # Especifica el host objetivo del escaneo


# Crear la carpeta para almacenar los resultados si no existe
folder_name = "temp"
os.makedirs(folder_name, exist_ok=True)

# Función para formatear el nombre del archivo basado en el tipo de escaneo y la fecha actual


def get_output_filename(scan_type):
    date_str = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
    return os.path.join(folder_name, f"nmap_output_{scan_type.strip('-')}_{date_str}.xml")


# Ejecutar los escaneos y guardar cada resultado en un archivo separado
for scan_type in scan_types:
    output_file = get_output_filename(scan_type)
    output_format = f"-oX {output_file}"

    # Formatear el comando usando f-strings
    command = f"nmap {scan_type} {target} {timing_template} {advanced_options} {verbosity_level} {output_format} {scan_service}"

    # Ejecutar el comando nmap
    result = subprocess.run(
        command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

    # Imprimir la salida y los errores
    print(f"Comando: {command}")
    print(result.stdout)
    if result.stderr:
        print("Error:", result.stderr)
```

#### Diferencias Potenciales en el Escaneo desde una Red Pública:

1. **Cantidad de Puertos y Servicios Detectados**: Es posible que algunos puertos o servicios que eran visibles internamente no estén accesibles o visibles desde una red pública debido a restricciones de firewall o políticas de enrutamiento que bloquean ciertos puertos o direcciones IP entrantes.

2. **Resultados de Latencia**: La latencia podría ser mayor cuando se escanea desde una red pública debido a la mayor distancia de red y los posibles obstáculos en el camino del tráfico de red, como firewalls y otros dispositivos de seguridad de red.

3. **Filtrado de Puertos**: Los resultados podrían mostrar un número mayor de puertos como filtrados o cerrados, lo que indica que hay medidas de seguridad adicionales (como firewalls) que protegen el servidor cuando se accede desde ubicaciones externas.

#### Implicaciones de Seguridad y Conclusiones:

1. **Eficacia de las Medidas de Perímetro**: Si el escaneo desde una red pública revela menos servicios o puertos abiertos, esto podría ser una señal de que las políticas de firewall y las configuraciones de seguridad perimetral están funcionando eficazmente para proteger el servidor contra accesos no autorizados desde fuera de la red interna.

2. **Exposición a Riesgos Externos**: Cualquier puerto o servicio que aún esté accesible desde una red pública debe ser evaluado críticamente para asegurarse de que está debidamente protegido, actualizado y monitoreado, ya que estos puntos de acceso son potencialmente explotables por actores de amenazas.

3. **Necesidad de Pruebas de Penetración Externas**: Realizar escaneos regulares desde fuera de la red interna es crucial para entender cómo un posible atacante podría ver el sistema. Esto ayuda a identificar y mitigar vulnerabilidades que podrían no ser aparentes desde el interior de la red.

4. **Configuración y Fortalecimiento de Firewall**: Los resultados pueden ayudar a informar sobre cómo configurar las reglas del firewall para equilibrar la accesibilidad y la seguridad. Puede ser necesario reforzar las reglas del firewall para bloquear o restringir el acceso a servicios sensibles que no necesitan ser accesibles globalmente.
